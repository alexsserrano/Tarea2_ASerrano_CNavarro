### Ejercicio 1:

> 1. Use filter to eliminate all words that are shorter than 4 letters from a list of words
`list='the notion of a lambda function goes all the way back to the origin of computer science'`

**Código de solución estudiante 1**

```python
import numpy as np

# Código de solución estudiante 1
texto = 'the notion of a lambda function goes all the way back to the origin of computer science'
words = texto.split()  # Divide la cadena en palabras

# Filtra las palabras que tienen menos de 4 letras
eliminate_words = list(filter(lambda word: len(word) <= 4, words))

print(eliminate_words)
```

**Código de solución estudiante 2**

```python
# Código de solución estudiante 2
texto = 'the notion of a lambda function goes all the way back to the origin of computer science'

words = texto.split()

# Definir función
def is_short(word):
    return len(word) < 4

# Filtrar
filtered_words = list(filter(is_short, words))

print("Palabras mas pequeñas a 4 letras:", filtered_words)
```

### Ejercicio 4:

> 4. Find all of the numbers from 1-1000 that are divisible by 7

**Código de solución estudiante 1**

```python

# Lista de números
number = np.arange(1, 1001)

# Resultado usando for tradicional
numbers_divisible_by_7_for = []
for num in number:
    if num % 7 == 0:
        numbers_divisible_by_7_for.append(num)
print("Resultado usando for tradicional:", numbers_divisible_by_7_for)

# Resultado usando compresión de listas
numbers_divisible_by_7_list_comprehension = [num for num in number if num % 7 == 0]
print("Resultado usando compresión de listas:", numbers_divisible_by_7_list_comprehension)

# Resultado usando la función filter
numbers_divisible_by_7_filter = list(filter(lambda num: num % 7 == 0, number))
print("Resultado usando la función filter:", numbers_divisible_by_7_filter)
```

### Ejercicio 5:

> 5. Find all of the numbers from 1-1000 that have a 3 in them

**Código de solución estudiante 1**

```python
# Lista de números
number = np.arange(1, 1001)

# Resultado usando for tradicional
numbers_with_3_for = []
for num in number:
    if '3' in str(num):
        numbers_with_3_for.append(num)
print("Resultado usando for tradicional:", numbers_with_3_for)

# Resultado usando compresión de listas
numbers_with_3_list_comprehension = [num for num in number if '3' in str(num)]
print("Resultado usando compresión de listas:", numbers_with_3_list_comprehension)

# Resultado usando la función filter
numbers_with_3_filter = list(filter(lambda num: '3' in str(num), number))
print("Resultado usando la función filter:", numbers_with_3_filter)
```

**Código de solución estudiante 2**

```python
# Lista de números
number = np.arange(1, 1001)

# Resultado usando for tradicional
numeros_con_3 = []
for number in range(1, 1001):
    if '3' in str(number):
        numeros_con_3.append(number)
print("Resultado usando for", numeros_con_3)


# Resultado usando compresión de listas
numeros_con_3 = [number for number in range(1, 1001) if '3' in str(number)]
print("Resultado usando comprensión de listas", numeros_con_3)

#Resultado usando filter
def tiene_3(number):
    return '3' in str(number)

numbers = list(range(1, 1001))
numbers_with_3 = list(filter(tiene_3, numbers))
print("Resultado usando filter", numbers_with_3)
```

### Ejercicio 6

>6. Imagine una rutina contable utilizada en una librería. Funciona en una lista con sublistas, que se ven así:

Escriba un programa en Python, que devuelve una lista con tuplas de tamaño 2. Cada tupla consiste en el número de pedido y el costo total del pedido. El producto debe aumentarse en 10€ si el valor de la orden es inferior a 100,00 €. 
Escribe un programa Python usando unicamente las funciones **lambda y map**.

```python
########### Solución
orders = [ ["34587", "Learning Python, Mark Lutz", 4, 40.95], 
           ["98762", "Programming Python, Mark Lutz", 5, 56.80], 
           ["77226", "Head First Python, Paul Barry", 3,32.95],
           ["88112", "Einführung in Python3, Bernd Klein", 3, 24.99]]
```

**Código de solución estudiante 1**

```python

# Función para calcular el costo total del pedido
total_cost = lambda order: order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0)

# Usamos map para aplicar la función a cada sublista y crear las tuplas
result = list(map(lambda order: (order[0], total_cost(order)), orders))

# Imprimir el resultado
for item in result:
    print(item)
```

**Código de solución estudiante 2**

```python
# Calcula el costo total por pedido y aplica un aumento de 10€ si es necesario
totales = list(map(lambda order: (order[0], order[2] * order[3] + 10 if order[2] * order[3] < 100 else order[2] * order[3]), orders))

print(totales)
```

### Ejercicio 7

>7. La misma librería, pero esta vez trabajamos en una lista diferente. Las sublistas de nuestras listas se ven así: 
[número de orden, (número de artículo, cantidad, precio por unidad), ... (número de artículo, cantidad, precio por unidad)] 

`orders = [[1, ("5464", 4, 9.99), ("8274",18,12.99), ("9744", 9, 44.95)], 
	       [2, ("5464", 9, 9.99), ("9744", 9, 44.95)],
	       [3, ("5464", 9, 9.99), ("88112", 11, 24.99)],
           [4, ("8732", 7, 11.99), ("7733",11,18.99), ("88112", 5, 39.95)] ]`

Escriba un programa que devuelva una lista de dos tuplas que tengan la información de (número de orden, cantidad total de pedido). Nuevamente, tenga en cuenta que si el pedido debe aumentarse en 10€ si el valor de la orden es inferior a 100,00 €.  Utilice la función `Reduce`.

**Código de solución estudiante 1**

```python
from functools import reduce

orders = [[1, ("5464", 4, 9.99), ("8274", 18, 12.99), ("9744", 9, 44.95)], 
          [2, ("5464", 9, 9.99), ("9744", 9, 44.95)],
          [3, ("5464", 9, 9.99), ("88112", 11, 24.99)],
          [4, ("8732", 7, 11.99), ("7733", 11, 18.99), ("88112", 5, 39.95)]]

def total_cost(order):
    return reduce(lambda x, y: x + y[1] * y[2], order[1:], 0)

def order_total(order):
    cost = total_cost(order)
    if cost < 100:
        cost += 10
    return order[0], cost

result = list(map(order_total, orders))

print(result)
```

```python
orders = [[1, ("5464", 4, 9.99), ("8274",18,12.99), ("9744", 9, 44.95)], 
           [2, ("5464", 9, 9.99), ("9744", 9, 44.95)],
           [3, ("5464", 9, 9.99), ("88112", 11, 24.99)],
           [4, ("8732", 7, 11.99), ("7733",11,18.99), ("88112", 5, 39.95)] ]
```

```python
from functools import reduce
def calculate_order_total(order):
    order_number = order[0]
    items = order[1:]
    total = reduce(lambda acumulado, item: acumulado + item[1] * item[2], items, 0)
    if total < 100:
        total += 10
    return (order_number, total)

totales = list(map(calculate_order_total, orders))

print(totales)
```
